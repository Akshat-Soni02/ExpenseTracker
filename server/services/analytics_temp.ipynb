```python
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
from prophet import Prophet # type: ignore
from statsforecast import StatsForecast # type: ignore
from statsforecast.models import AutoARIMA, AutoETS # type: ignore
# from statsforecast import ETS  # type: ignore
```

```python
# Load expense data
file_path = "sampledataset.csv"  # Update with your file path
df = pd.read_csv(file_path)
df
```

```python
# Convert 'date' column to datetime format
df["date"] = pd.to_datetime(df["date"])
df
```

```python
# Aggregate daily expenses into monthly totals
df["month"] = df["date"].dt.to_period("M")  # Convert to monthly period
monthly_expense = df.groupby("month")["amount"].sum().reset_index()
monthly_expense
```

```python
# Convert 'month' back to datetime format for Prophet
monthly_expense["month"] = monthly_expense["month"].astype(str) + "-01"
monthly_expense["month"] = pd.to_datetime(monthly_expense["month"])
monthly_expense["month"]
```

```python
# Rename columns for Prophet
monthly_expense = monthly_expense.rename(columns={"month": "ds", "amount": "y"})
monthly_expense
```

```python
monthly_expense["unique_id"] = "expenses"
```

```python
# Initialize and fit Prophet model
# model = Prophet()
# model.fit(monthly_expense)

sf = StatsForecast(
    models=[AutoARIMA(), AutoETS(season_length=12)],  # AutoARIMA & ETS models
    freq="MS",  # Monthly frequency
    n_jobs=-1  # Use all CPU cores
)
```

```python
forecast = sf.forecast(df=monthly_expense, h=24)
```

```python
# Reset index to ensure we have a datetime index
forecast = forecast.reset_index()

# Extract next month's prediction correctly
next_month = forecast.iloc[10]  # First row is next month's prediction

# Ensure the index is treated as a datetime
predicted_date = pd.to_datetime(next_month["ds"])  # Assuming "ds" is the date column

# Print the prediction with the correct format
print(f"Predicted Expense for {predicted_date.strftime('%B %Y')}: {next_month['AutoARIMA']:.2f}")
```

```python
plt.figure(figsize=(10, 5))

# Plot only available months
plt.plot(monthly_expense["ds"], monthly_expense["y"], label="Actual Expenses", marker="o")

# Ensure forecast dates start from the last available month
forecast_filtered = forecast[forecast["ds"] >= monthly_expense["ds"].min()]  # Filter out invalid dates

plt.plot(forecast_filtered["ds"], forecast_filtered["AutoARIMA"], label="Predicted Expenses (AutoARIMA)", linestyle="dashed")

# Mark today's date (latest available data point)
plt.axvline(x=monthly_expense["ds"].max(), color="red", linestyle="--", label="Today")

plt.legend()
plt.xlabel("Date")
plt.ylabel("Total Monthly Expense")
plt.title("Monthly Expense Forecast with AutoARIMA")
plt.show()
```
